# A/B Test experiment for conversions

#### Import AB_experiment

```python
from byas_ab import AB_experiment
```

```python
all_ctrl, target_ctrl = 16500, 30
all_test, target_test = 17000, 50

exp1 = AB_experiment(all_ctrl, target_ctrl,               # create exptriment
                     all_test, target_test, "Experiment 1")       
exp1.plot_distrib([-0.002, 0.01])                         # plot distribution
print(exp1)
```

# Make serias experiments

```python
experiments = []
experiments.append(exp1)
```

```python
exp2 = AB_experiment(all_ctrl+1000, target_ctrl+5,               # create exptriment 2
                     all_test+1000, target_test+10, "Experiment 2")
experiments.append(exp2)

exp2.plot_distrib()
print(exp2)
```

```python
exp3 = AB_experiment(all_ctrl+1000, target_ctrl+5,               # create exptriment 2
                     all_test+1000, target_test+10, "Experiment 3")
experiments.append(exp3)
```

## a few experiments

```python
import matplotlib.pyplot as plt


for ex in experiments:
    print(ex)
    
xs = [i for i in range(len(experiments))]

probs = [x.probability for x in experiments]
lifts = [x.lift for x in experiments]

plt.plot(xs, lifts, 'go-', label=f"lifts on Experiment")
plt.plot(xs, probs, 'ro-', label=f"probabilitys on Experiment")
plt.grid()
plt.legend()
plt.show
```
